# Calculating Opportunity score

Opportunity Score: Total commercial vehicles x (1-penetration) x GDP forecast

```python
result_df = duckdb.sql(f"""
SELECT *,
ROUND((total_commercial_vehicles) * (100 - telematics_penetration_pct) * (gdp_growth_forecast),2) as opportunity_score
FROM read_csv_auto('market_ref.csv')
ORDER BY OPPORTUNITY_SCORE DESC;
""").df()
```
